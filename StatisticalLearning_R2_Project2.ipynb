```python
 ##Answer the following questions with statistical evidence 
##a. Do charges of people who smoke differ significantly from the people who don't? 
```

```python
##using T test to find the effect of smoking on charges

## Null hypothesis : There is no significant different in charges between people who smoke and people who does not.(charges of people who smoke and who does not smoke are same.)
## Alternate hypothesis : There is significant different in charges between people who smoke and people who does not.(charges of people who smoke and who does not smoke are different.
## replacing the smoker with original data
ins['smoker'] = ins['smoker'].replace({1:'yes', 0:'no'})
```

```python
## selecting charges corresponding to smoker as an array
data1 = np.array(ins[ins['smoker'] == 'yes'].charges)

## selecting charges corresponding to non-smoker as an array
data2 = np.array(ins[ins['smoker'] == 'no'].charges)
```

```python
##performing independent t test , considering significant level(alpha) as 5 %
t, p_value = stats.ttest_ind(data1 ,data2 , axis = 0)
```

```python
if p_value < 0.05:
    print(f'Rejecting null as p_value {p_value} is lower than significant level, hence alternate hypothesis is true')
    print(f'Charges of smoker and non smoker are not  same as the p_value {p_value} is less than 0.05')
else:
    print(f'Failed to reject null as p_value {p_value} is larger than significant level, hence null is accepted')
    print(f'charges of smoker and non smoker are  same as the p_value {p_value} is greater than 0.05')
```

```python
 ##Does bmi of males differ significantly from that of  females?  
```

```python
## using t test to find the depedency of BMI on gender
```

```python
## Null hypothesis :  No change in BMI of male and female. ( Gender has no effect on BMI)
## Alternate Hypothesis : There is significant difference between BMI of male and female. (Gender has an effect on BMI )

## replacing the sex with original data
ins['sex'] = ins['sex'].replace({0:'male', 1:'female'})
```

```python
## selecting bmi corresponding to male as an array
datam = np.array(ins[ins['sex'] == 'male'].bmi)

## selecting bmi corresponding to female as an array

dataf = np.array(ins[ins['sex'] == 'female'].bmi)
```

```python
## using independent T-test to find the p_value / Setting significant level as 5 %

t , p_value = stats. ttest_ind(datam , dataf , axis = 0)

if p_value < 0.05:
    print(f'Null is rejected due to low p_value {p_value}, hence alternate hypthesis is true')
    print(f'Gender has an effect on bmi as the p_value {p_value} is significantly lower than alpha of 0.05')
else:
    print(f'Failed to reject null due to higher p_value {p_value},hence null is accepted')
    print(f'Gender has no effect on bmi as the p_value {p_value} is significantly higher than alpha of 0.05')
```

```python
##c. Is the proportion of smokers significantly different  in different genders? 
```

```python
## chi-square test is used for this use case as two categorical data are involved in it
```

```python
## Null hypothesis : Gender has no effect on smoking habits
## Alternate hypothesis : Gender has an effect on smoking habits

## forming contigency table for two categorical data

crosstab = pd.crosstab(ins['sex'],ins['smoker'])

chi, p_value ,dof, expected = stats.chi2_contingency(crosstab)
if p_value <  0.05:
    print(f'Null is rejected as the p_value {p_value} is lower than significant level alpha,hence alternate hypothesis is true')
    print(f'Gender has an effect on smoking habbits as the p_value {p_value} is lower than significant level of 0.05')
else:
    print(f'Failed to rejected as the p_value {p_value}is greater than significant level alpha,hence accepting null')
    print(f'Gender has no effect on smoking habbits as the p_value {p_value} is greater than significant level of 0.05')

crosstab
```

```python
##d. Is the distribution of bmi across women with no children, one child and two children, the same ? 
```

```python
## One way Anova is performed as this data includes comparison between one independent variable and two or more categorical value
```

```python
##Null Hypothesis:  No of children has no effect on bmi of women
##Alternate hypothesis : No of Children has an effect on bmi of women 
```

```python

df = ins[ins['sex'] == 'female']
```

```python
dfzero  = df[df.children == 0]['bmi']
dfone = df[df.children == 1]['bmi']
dftwo = df[df.children == 2]['bmi']
```

```python
stat , p_value = stats.f_oneway(dfzero, dfone,dftwo)
```

```python
# Setting significance level at 5%
```

```python
if p_value < 0.05:  
    print(f'Null is rejected as the p_value {p_value} is lower than significance level,hence alternetate hypothesis is true')
    
    print(f' No of Children has an effect on bmi of women  as the p_value ({p_value}) < 0.05')
else:
    print(f' failed to  reject null as the p_value {p_value} is gretaer  than significance level,hence accepting null')
    print(f' No of children has no effect on bmi of women as the p_value ({p_value}) > 0.05')
```
